```python
import numpy as np
import matplotlib.pyplot as plt

# Define the quadratic function
def f(x):
    a, b, c = 1, -3, 2  # You can change these
    return a*x**2 + b*x + c 

# Define the cost function (absolute value)
def cost(x):
    return abs(f(x))

# Hill Climbing Algorithm
def hill_climb(start_x, step_size=0.125, max_iterations=200):
    x = start_x
    history = [(x, cost(x))]

    for _ in range(max_iterations):
        neighbors = [x + step_size, x - step_size]
        best_neighbor = min(neighbors, key=cost)

        if cost(best_neighbor) < cost(x):
            x = best_neighbor
            history.append((x, cost(x)))
        else:
            break  # Local minima reached

    return x, history

# Run the algorithm
# start_x = np.random.uniform(-10, 10)
root, path = hill_climb(-1)

# Extract x and y values for plotting
x_vals, y_vals = zip(*path)

# Plotting
x_range = np.linspace(-1, 4, 400)
y_range = [f(x) for x in x_range]

plt.figure(figsize=(10, 6))
plt.plot(x_range, y_range, label="f(x) = x - 3x + 2", color='blue')
plt.scatter(x_vals, [f(x) for x in x_vals], color='red', label='Hill Climb Steps')
plt.title("Hill Climbing to Find Root of Quadratic Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='gray', linestyle='--')
plt.legend()
plt.grid(True)
plt.show()

print(f"Found approximate root at x = {root:.4f}, f(x) = {f(root):.4e}")
```
